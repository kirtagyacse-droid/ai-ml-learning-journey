```python
stats = np.array([[1,2,3],[4,5,6]])
print(np.min(stats))
print(np.max(stats, axis=1))
print(np.max(stats, axis=0))
```

```python
print(np.sum(stats, axis=0)) #columnwise sum
print(np.sum(stats, axis=1)) #rowsise sum
```

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
after = before.reshape((8,1))
print(after)

```

```python
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])
print(np.vstack([v1,v2]))
print(np.hstack([v1,v2]))
```

```python
data = np.array([[1,50,3],[80,5,90],[10,20,100]])
print(data[data>50])
print(np.any(data>90, axis=0))
print(np.any(data>0, axis=0))
```
